```python
import json
from datetime import datetime
```

```python
with open("drug-ndc-0001-of-0001.json") as f:
    data = json.load(f)
    
```

I am checking how many items is in the file

```python
len(data["results"])
```

```python
list(filter(lambda x: 'generic_name' in x and x['generic_name'] == "Piroxicam", data["results"]))
```

```python
manufacturers = data['results']
manufacturers = filter(lambda x: 'manufacturer_name' in x['openfda'], manufacturers )
manufacturers = list(map(lambda x: x['openfda']['manufacturer_name'][0], manufacturers))
```

```python
from collections import Counter

c = Counter(manufacturers)

```

```python
len(c.most_common())
```

Teraz będziemy odpowiadali na pytanie jakie leki są najbardziej wpływowe.
Zeby to zobaczyć to ciekawe jest czy sa jakies powiąznania miedzy tymi lekami.
Ponizej mam liste leków ktore maja taki sam generic name. Czy posortowanie ich
w kolejnosci start marketing date da mi odpowiedz.

```python
list(
    sorted(
        map(
            lambda x: 
                (
                    x["generic_name"],
                    datetime.strptime(x["marketing_start_date"], '%Y%m%d'),
                    datetime.strptime(x["listing_expiration_date"], '%Y%m%d'),
                    x["labeler_name"],
                  
                )
            ,
            filter(
                lambda x: "generic_name" in x and x["generic_name"] == "Piroxicam",
                data["results"],
            ),
        ),
        key= lambda x: (x[1])
    )
)
```

```python
list(
    filter(
                lambda x: "brand_name" in x and x["brand_name"] == "Glucosamine Chondroitin",
                data["results"],
            )
)
```

Kolejnym pytaniem czy są leki które nie są generykami, czyli takie które nie mają generic name

```python

list(filter(
                lambda x: "brand_name" not in x,
                data["results"],
            )
)[0]
```

Jaki generyk jest najbardziej popularny?

```python
generic_counter = Counter(list(
    map( lambda x: x['generic_name'],
        filter(
                    lambda x: "generic_name" in x,
                    data["results"]
                )
    )
))
```

```python
generic_counter.most_common()
```
