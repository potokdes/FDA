```python
with open("../data/drug-ndc-0001-of-0001.json") as f:
    data = json.load(f)
    
```

```python
manufacturers = data['results']
manufacturers = filter(lambda x: 'manufacturer_name' in x['openfda'], manufacturers )
manufacturers = list(map(lambda x: x['openfda']['manufacturer_name'][0], manufacturers))
```

```python
from collections import Counter

c = Counter(manufacturers)

```

```python
len(c.most_common())
```

Teraz bdziemy odpowiadali na pytanie jakie leki s najbardziej wpywowe.
Zeby to zobaczy to ciekawe jest czy sa jakies powiznania miedzy tymi lekami.
Ponizej mam liste lekw ktore maja taki sam generic name. Czy posortowanie ich
w kolejnosci start marketing date da mi odpowiedz.

```python
list(
    sorted(
        map(
            lambda x: 
                (
                    x["generic_name"],
                    datetime.strptime(x["marketing_start_date"], '%Y%m%d'),
                    datetime.strptime(x["listing_expiration_date"], '%Y%m%d'),
                    x["labeler_name"],
                  
                )
            ,
            filter(
                lambda x: "generic_name" in x and x["generic_name"] == "Piroxicam",
                data["results"],
            ),
        ),
        key= lambda x: (x[1])
    )
)
```

```python
list(
    filter(
                lambda x: "brand_name" in x and x["brand_name"] == "Glucosamine Chondroitin",
                data["results"],
            )
)
```

Kolejnym pytaniem czy s leki ktre nie s generykami, czyli takie ktre nie maj generic name

```python

list(filter(
                lambda x: "brand_name" not in x,
                data["results"],
            )
)[0]
```

Jaki generyk jest najbardziej popularny?

```python
generic_counter = Counter(list(
    map( lambda x: x['generic_name'],
        filter(
                    lambda x: "generic_name" in x,
                    data["results"]
                )
    )
))
```

```python
generic_counter.most_common()
```
